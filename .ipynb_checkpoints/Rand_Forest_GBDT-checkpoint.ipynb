```python
import warnings
warnings.filterwarnings('ignore')

import yfinance as yf
import pandas as pd
import numpy as np
import altair as alt
import multi_model_builder as mmb
import project_functions2 as pf
import multi_regressor_builder as mrb

from matplotlib import pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
alt.renderers.enable('default')
alt.data_transformers.enable('json')
```

```python
googl = yf.Ticker('GOOGL')
amzn = yf.Ticker('AMZN')
aapl = yf.Ticker('AAPL')
fb = yf.Ticker('FB')
msft = yf.Ticker('MSFT')
tsla = yf.Ticker('TSLA')
```

```python
aapl_df = aapl.history(period='max')
amzn_df = amzn.history(period='max')
fb_df = fb.history(period='max')
googl_df = googl.history(period='max')
msft_df = msft.history(period='max')
tsla_df = tsla.history(period='max')
```

```python
stock_dfs = [aapl_df, amzn_df, fb_df, googl_df, msft_df, tsla_df]

for stock_df in stock_dfs:
    stock_df = pf.date_time_prep(stock_df) 
    stock_df = pf.rolling_aves(stock_df)
    stock_df = pf.future_close_setup(stock_df, 1)
    stock_df.drop(stock_df.tail(1).index,inplace=True)
```

```python
def combiner(stock_dfs):
    combine_df = stock_dfs[0]
    i = 1
    while i < len(stock_dfs):
        combine_df = pd.concat([combine_df, stock_dfs[i]])
        i += 1
    
    combine_df.sort_values(by=['Date'], inplace=True)
    return combine_df
```

```python
combine_df = combiner(stock_dfs)
combine_df.tail()
```

```python
rand_forest_df = mrb.train_multiple_random_forests(combine_df, 365, stock_dfs, 10, 200, 5, 20)
```

```python
gbdt_df = mrb.train_multiple_boosted_DTs(combine_df, 365, stock_dfs, 3, 20, 5, 20)
```

```python
def column_shrink(combine_df):
    shrink_combine_df = combine_df.drop([
        'Open', 'High', 'Low', 'Volume', 'Dividends', 'Stock Splits',
       'Day', 'Month', 'Year', 'Days From IPO',
       '5 Day Open Mean', '5 Day High Mean', '5 Day Low Mean',
       '5 Day Close Mean', '5 Day Volume Mean', '5 Day Open Var',
       '5 Day High Var', '5 Day Low Var', '5 Day Close Var',
       '5 Day Volume Var', '10 Day Open Mean', '10 Day High Mean', '10 Day Low Mean',
       '10 Day Close Mean', '10 Day Volume Mean', '10 Day Open Var',
       '10 Day High Var', '10 Day Low Var', '10 Day Close Var',
       '10 Day Volume Var', '20 Day Open Mean', '20 Day High Mean', '20 Day Low Mean',
       '20 Day Close Mean', '20 Day Volume Mean', '20 Day Open Var',
       '20 Day High Var', '20 Day Low Var', '20 Day Close Var',
       '20 Day Volume Var', '20 Day High', '20 Day Low'
    ], axis=1)
    return shrink_combine_df
```

```python
shrink_combine_df = column_shrink(combine_df)
```
